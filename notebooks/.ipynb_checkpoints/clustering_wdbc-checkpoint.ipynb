# K-means — demo WDBC (clustering)

Clustering **non supervisionato** sul dataset Breast Cancer Wisconsin con il nostro `KMeans` scritto **da zero** (solo numpy). Mettiamo da parte le etichette durante il clustering e le riusiamo **solo** alla fine, per valutare quanto i cluster somiglino alla diagnosi reale. Coerente con le slide *14.Clustering*.

## 1. Setup e caricamento dati

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from src.kmeans import KMeans
from src.preprocessing import StandardScaler
from src.metrics import purity, silhouette, confusion_matrix

df = pd.read_csv('../data/raw/wdbc.csv')
feat = [c for c in df.columns if c not in ('target', 'diagnosis')]
X = df[feat].values            # 30 feature
y = df['target'].values        # 0=maligno, 1=benigno -- SOLO per valutazione finale
print('shape:', X.shape)
df['diagnosis'].value_counts()
```

## 2. Standardizzazione (qui serve davvero)

K-means assegna i punti in base alla **distanza euclidea**: senza standardizzare, le feature con valori grandi (es. `area` ~1000) dominerebbero quelle piccole. Quindi standardizziamo — al contrario del Naïve Bayes, che è invariante alla scala.

```python
Xs = StandardScaler().fit_transform(X)
print('media ~0, std ~1 dopo scaling:', np.round(Xs.mean(), 3), np.round(Xs.std(), 3))
```

## 3. Scelta di K: metodo del gomito

Eseguiamo k-means per K crescenti e guardiamo la **distorsione** $J$ (slide 29). Il "gomito" è il punto oltre il quale aggiungere cluster non riduce più di molto $J$.

```python
Ks = range(1, 9)
inertie = [KMeans(n_clusters=k, n_init=10, random_state=42).fit(Xs).inertia_ for k in Ks]
plt.figure(figsize=(6, 4))
plt.plot(list(Ks), inertie, '-o')
plt.axvline(2, ls='--', color='tab:red', alpha=0.7)
plt.xlabel('numero di cluster K'); plt.ylabel(r'distorsione media $J$')
plt.title('Metodo del gomito'); plt.grid(True, alpha=0.3); plt.show()
print('Il gomito e\' intorno a K=2, coerente con le due diagnosi (maligno/benigno).')
```

## 4. K-means con K = 2

Inizializzazione **k-means++** e 10 restart (si tiene la distorsione minima).

```python
km = KMeans(n_clusters=2, init='k-means++', n_init=10, random_state=42).fit(Xs)
print('distorsione J (media): %.4f' % km.inertia_)
print('iterazioni del miglior restart:', km.n_iter_)
print('dimensione dei cluster:', np.bincount(km.labels_))
```

## 5. Confronto cluster vs diagnosi (validazione esterna)

Ora riusiamo le etichette: quanto i cluster trovati **senza supervisione** somigliano alla diagnosi reale? (slide 67-69)

```python
C = confusion_matrix(km.labels_, y, labels=[0, 1])   # righe = cluster, colonne = classe
print('righe = cluster, colonne = classe (0=maligno, 1=benigno):')
print(C)
print('\npurity = %.4f' % purity(y, km.labels_))
print('-> ogni cluster e\' assegnato alla classe di maggioranza; la purity e\' la frazione corretta.')
```

## 6. Silhouette (validazione interna)

Senza usare le etichette: quanto sono **compatti e separati** i cluster? (slide 73-76). Vicino a 1 = ottimo, vicino a 0 = cluster sovrapposti.

```python
print('silhouette media: %.4f' % silhouette(Xs, km.labels_))
```

```python
# visualizzazione su 2 feature: cluster trovati vs diagnosi reale
d = np.abs(km.cluster_centers_[0] - km.cluster_centers_[1])
a, b = np.argsort(d)[::-1][:2]                         # 2 feature piu' 'definenti'
fig, ax = plt.subplots(1, 2, figsize=(11, 4.5), sharex=True, sharey=True)
ax[0].scatter(Xs[:, a], Xs[:, b], c=km.labels_, cmap='coolwarm', s=10, alpha=0.6)
ax[0].set_title('cluster trovati da k-means')
ax[1].scatter(Xs[:, a], Xs[:, b], c=y, cmap='coolwarm', s=10, alpha=0.6)
ax[1].set_title('diagnosi reale')
for a_ in ax: a_.set_xlabel(feat[a]); a_.grid(True, alpha=0.3)
ax[0].set_ylabel(feat[b]); plt.tight_layout(); plt.show()
```

## 7. Confronto con sklearn

Riferimento esterno per validare la nostra implementazione (la numerazione dei cluster è arbitraria, quindi confrontiamo **distorsione** e **silhouette**, non le etichette).

```python
from sklearn.cluster import KMeans as SKKMeans
from sklearn.metrics import silhouette_score
sk = SKKMeans(n_clusters=2, n_init=10, random_state=0).fit(Xs)
print('distorsione (somma)  noi: %.1f   sklearn: %.1f' % (km.inertia_*len(Xs), sk.inertia_))
print('silhouette           noi: %.4f   sklearn: %.4f' % (silhouette(Xs, km.labels_), silhouette_score(Xs, sk.labels_)))
print('purity               noi: %.4f   sklearn: %.4f' % (purity(y, km.labels_), purity(y, sk.labels_)))
```

## 8. Conclusioni

- K-means **da zero** (multi-restart + k-means++): ciclo assign/update, distorsione minimizzata.
- Il **gomito** conferma K=2, coerente con le due diagnosi.
- Senza vedere le etichette, i cluster ritrovano la diagnosi con **purity ≈ 0.91** (validazione esterna); la **silhouette ≈ 0.34** dice che i due gruppi si sovrappongono un po' (validazione interna).
- **Distorsione identica a sklearn** → implementazione corretta.
- A differenza del Naïve Bayes, qui la **standardizzazione è essenziale** perché k-means si basa sulle distanze.
